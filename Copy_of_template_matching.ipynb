## Multi-Template Matching
multi_template_matching_output.png


```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/test_1 (1).png")
template = cv2.imread('/content/template_1 (1).png')
(tH, tW) = template.shape[:2]
# # display the  image and template to our screen
print("Test Image")
cv2_imshow(image)
print("template image")
cv2_imshow(template)
```

```python
template.shape[:2]
```

```python
# convert both the image and template to grayscale
imageGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
templateGray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
# perform template matching
print("[INFO]performing template matching...")
result = cv2.matchTemplate(imageGray, templateGray,	cv2.TM_SQDIFF_NORMED)
```

```python
threshold=0.0001

(yCoords, xCoords) = np.where(result <= threshold)
clone = image.copy()
print("matched locationS")
# loop over our starting (x, y)-coordinates
for (x, y) in zip(xCoords, yCoords):
	# draw the bounding box on the image
	cv2.rectangle(clone, (x, y), (x + tW, y + tH), (0, 0, 255))
cv2_imshow(clone)
```
